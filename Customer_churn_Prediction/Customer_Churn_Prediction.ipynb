```python
df["Churn"] = df["Churn"].astype(int)

Y_train = df["Churn"]
X_train = df.drop(labels = ["Churn"],axis = 1)
```

```python
from sklearn.model_selection import  cross_val_score,GridSearchCV,RandomizedSearchCV
from sklearn.metrics import roc_auc_score


def CalculateScore(clf, X_train, Y_train):
    
    # 10 Folds Cross Validation 
    clf_score = cross_val_score(clf, X_train, Y_train, cv=10)
    clf_score.mean()
  
    
    print( "Model Result\n" +
           "Accuracy of " + str(clf.__class__.__name__) + " :\n "
           "\nCross Val Score : " + str(clf_score.mean())) 
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=15)
tree.fit(X_train, Y_train)
```

```python
CalculateScore(tree, X_train, Y_train)
```
